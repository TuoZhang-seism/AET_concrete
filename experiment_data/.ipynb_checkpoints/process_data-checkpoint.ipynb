```python
############################get packages and define functions#################
import obspy
import obspy.signal.filter
import obspy.signal.invsim
import os
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates 
from scipy import optimize as op
from scipy import signal
import re
from obspy import Trace, Stream
from obspy.core import UTCDateTime
import obspy.signal.freqattributes
from os import path
import glob
import matplotlib.colors as colors
import matplotlib.animation as animation
import matplotlib.ticker
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline


def local_energy_density(x,y,z,mc,rec_type):
        lenth = 0
        ar = np.arange(-lenth,1+lenth,1)
        s_ar = mc.head['GridSpacing'][0] * mc.head['GridSpacing'][0]
        x_ind = np.argmin(np.abs(mc.axes['x']-x))
        y_ind = np.argmin(np.abs(mc.axes['y']-y))
        z_ind = np.argmin(np.abs(mc.axes['z']-z))
#       time_ind=(mc.axes['t']*10).astype(int)
        x_ar = ar+ x_ind
        x_ar = [n for n in x_ar if n > 0 and n <len(mc.axes['x'])-1]
        y_ar = ar+ y_ind
        y_ar = [n for n in y_ar if n > 0 and n <len(mc.axes['y'])-1]
        data_ini=mc.data[rec_type][:,:,z_ind,:]
        data_sed = data_ini[x_ar,:,:]
        data_thi = data_sed[:,y_ar,:]
        #data = (np.mean(np.mean(data_thi,0),0)/s_ar)/mc.head['npart']
        data = np.sum(np.mean(np.mean(data_thi,0),0),0)/s_ar/mc.head['npart']/mc.head['emodulo']/10
        #data = np.mean(np.mean(np.mean(data_thi,0),0),0)/s_ar
        return data,x_ind,y_ind
            

def position(Number): 
    r = re.compile("([a-zA-Z]+)([0-9]+)")
    m = r.match(Number)
    num = int(m.group(2)) 
    with open('Sensor_coordinates.txt') as f:
        lines = f.readlines()
    line = lines[num-119]
    line =line.replace("  "," ")
    line =line.replace("  "," ")
    line =line.replace("  "," ")
    y = int(line.split(' ')[5])
    x = int(line.split(' ')[4])
    return x,y

def read_file(file_name,folder): 
    tr = Trace()
    tr.data = np.fromfile(folder+'/'+file_name,'>i2')
    tr.stats.station = file_name.split('-')[4].split('_')[3]
    tr.stats.location = file_name.split('-')[4].split('_')[2]
    tr.stats.sampling_rate = 2e6
    tr.stats.delta = 5e-7
    tr.stats.npts = 10000
    tr.stats.network =file_name.split('-')[4].split('_')[4].split('.')[0]
    Year = int(file_name.split('-')[0])
    Month = int(file_name.split('-')[1])
    Day = int(file_name.split('-')[2].split('_')[0])
    Hour = int(file_name.split('-')[2].split('_')[1])
    Minute = int(file_name.split('-')[3])
    Second = int(file_name.split('-')[4].split('_')[0])        
    tr.stats.starttime = UTCDateTime(Year, Month, Day, Hour, Minute, Second)
    return tr
```

# Input original data

```python
ori_input = []
for i in [1,2,3]:
    path28 = './GFZ_DATA_28_10_2020/Meas'+str(int(i))
    for dirs in os.listdir(path28):
        ori_input.append(read_file(dirs,path28))

for i in [1,2]:
    path29 = './GFZ_DATA_29_10_2020/Meas'+str(int(i))
    for dirs in os.listdir(path29):
        ori_input.append(read_file(dirs,path29))

for i in [1,2]:
    path30 = './GFZ_DATA_30_10_2020/Meas'+str(int(i))
    for dirs in os.listdir(path30):
        ori_input.append(read_file(dirs,path30))

SampFreq =  2e6
dt = 1 /SampFreq *1000 #ms
NumSamp=  10000
t = np.arange(0,ori_input[0].stats.delta*ori_input[0].stats.npts,ori_input[0].stats.delta)*1000 #ms
```

# Process the data

```python
t_tri = int(0.2/dt)
#####################the 1st step: detrend#############
detrend_D = []
for i in np.arange(len(ori_input)):
    tem_DATA = Trace()
    tem_DATA.stats = ori_input[i].stats
    tem_DATA.data = np.copy(ori_input[i].data)
    tem_DATA.data = tem_DATA.data - np.mean(ori_input[i].data[0:180])
    detrend_D.append(tem_DATA)

#####################the 2nd step: remove crosstalk 0 ms - 0.2 ms setted by 0 #############
re_cro_D = []
Source_or = []
for i in np.arange(len(detrend_D)):
    tem_DATA = Trace()
    tem_DATA.stats = detrend_D[i].stats
    S_tem = Trace()
    S_tem.stats = detrend_D[i].stats
    tem_DATA.data = np.copy(detrend_D[i].data)
    distance = (position(detrend_D[i].stats.location)[0]-position(detrend_D[i].stats.station)[0])**2+ (position(detrend_D[i].stats.location)[1]-position(detrend_D[i].stats.station)[1])**2
    distance = np.sqrt(distance)/1000
    t_arr = int(distance/4.475/dt)
    S_tem.data = np.copy(tem_DATA.data[0:t_tri])
    tem_DATA.data[0:t_arr] = np.mean(tem_DATA.data[t_arr:t_arr+200])   
    tem_DATA.data = tem_DATA.data -  np.mean(tem_DATA.data[t_arr:t_arr+200])
    re_cro_D.append(tem_DATA)
    Source_or.append(S_tem)
    
#####################the 3rd step: fillter#############
freq_min = 60e3
freq_max = 120e3
filltered_D  = []
filltered_S  = []
for i in np.arange(len(re_cro_D)):
    tem_DATA = Trace()
    tem_DATA.stats = re_cro_D[i].stats
    tem_DATA.data = np.copy(re_cro_D[i].data)
    tem_DATA.taper(0.05)
    tem_DATA.filter(type='bandpass',freqmin= freq_min, freqmax= freq_max)
    filltered_D.append(tem_DATA)
    ###
    S_tem = Trace()
    S_tem.stats = Source_or[i].stats
    S_tem.data = np.copy(Source_or[i].data)
    S_tem.taper(0.05)
    S_tem.filter(type='bandpass',freqmin= freq_min, freqmax= freq_max)
    filltered_S.append(S_tem)

#####################the 4th step: calculate the envelope#############
envelope_D  = []
envelope_S  = []
for i in np.arange(len(filltered_D)):
    tem_DATA = Trace()
    tem_DATA.stats = filltered_D[i].stats
    tem_DATA.data = np.copy(filltered_D[i].data)
    tem_DATA.envelope = obspy.signal.filter.envelope(tem_DATA.data)
    envelope_D.append(tem_DATA)
    ######
    S_tem = Trace()
    S_tem.stats = filltered_S[i].stats
    S_tem.data = np.copy(filltered_S[i].data)
    S_tem.envelope = obspy.signal.filter.envelope(S_tem.data)
    envelope_S.append(S_tem)
    
#####################the 5th step: average repeated recording#############
##each transducer combination has 7 records within two days at the same time
average_D = []
average_S = []
for source in np.arange(19):
    for station in np.arange(19):
        tem_DATA = Trace()
        S_tem = Trace()
        y_data = []
        y_env = []
        S_data = []
        S_env = []
        if source != station:
            for i in np.arange(len(envelope_D)):
                if envelope_D[i].stats.location == 'T0'+str(int(source+119)) and envelope_D[i].stats.station == 'R0'+str(int(station+119)):
                    tem_DATA.stats = envelope_D[i].stats
                    y_data.append(envelope_D[i].data)  
                    y_env.append(envelope_D[i].envelope)  
                    S_data.append(envelope_S[i].data)  
                    S_env.append(envelope_S[i].envelope)
            tem_DATA.data = np.mean(y_data,0) 
            tem_DATA.envelope = np.mean(y_env,0) 
            average_D.append(tem_DATA)
            ######
            S_tem.data = np.mean(S_data,0) 
            S_tem.envelope = np.mean(S_env,0) 
            average_S.append(S_tem)

####################calculate the sensor coupling ##################################
eff_from_sou_rec = np.zeros((19,19))
xxt = []
for source in np.arange(19):
    xxt.append(str(int(source+119)))
    for station in np.arange(19):
        if source != station:
            for i in np.arange(len(average_D)):
                if average_D[i].stats.location == 'T0'+str(int(source+119)) and average_D[i].stats.station == 'R0'+str(int(station+119)):
                    eff_from_sou_rec[source,station]= np.average(average_D[i].envelope[int(3.8/dt):int(4.8/dt)])
        else:
            eff_from_sou_rec[source][station]=1
eff_from_station = np.exp(np.sum(np.log(eff_from_sou_rec),0)/18)
eff_from_source  = np.exp(np.sum(np.log(eff_from_sou_rec),1)/18)
#####################the 6th step: average repeated recording#############
processed_D = []
for source in np.arange(19):
    for station in np.arange(19):
        tem_DATA = Trace()
        if source != station:
            for i in np.arange(len(average_D)):
                if average_D[i].stats.location == 'T0'+str(int(source+119)) and average_D[i].stats.station == 'R0'+str(int(station+119)):
                    tem_DATA.stats = average_D[i].stats
                    tem_DATA.data =  average_D[i].data
                    tem_DATA.envelope = average_D[i].envelope/eff_from_source[source]/eff_from_station[station]
                    processed_D.append(tem_DATA)
```
